```python
df=pd.read_csv('imdb_data (1).csv')
```

```python
df['revenue']=pd.to_numeric(df['revenue'])
df['budget']=pd.to_numeric(df['budget'])
df['profit']=df['revenue']-df['budget']
df.profit
```

# The movie which made the highest profit is:

```python
highest_profit_movies=df.loc[df['profit'].idxmax()]
movie_name=highest_profit_movies['original_title']
print(movie_name)
df.profit.max()
```

```python
profit=highest_profit_movies['profit']
crew=highest_profit_movies['crew']
print(crew)
print(profit)
```

```python
import pandas as pd
df=pd.read_csv('imdb_data (1).csv')
df['ROI']=(df['revenue']-df['budget'])/df['budget']
average_roi_by_language=df.groupby('original_language')['ROI'].mean()
highest_roi_language=average_roi_by_language.idxmax()
highest_roi_value=average_roi_by_language.max()
print(highest_roi_language)
print(highest_roi_value)
average_roi_by_language
```

```python
unique_geners=df['genres'].unique()
unique_geners
```

# Name of the director and producer of the movie which made highest profit are:
